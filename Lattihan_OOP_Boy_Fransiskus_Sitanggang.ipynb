2. Buat sebuah program OOP dengan ketentuan:

  - berbeda dari contoh program yang ada di materi (ppt dan colab)
  - memiliki 1 parent class
  - memiliki setidaknya 2 child class
  - menerapkan konsep dasar OOP (kecuali abstraction)
  - dilengkapi dengan docstring, contoh docstring bisa dilihat dalam program hero.py

  Setelah program selesai, buatlah:
  - setidaknya 2 instance untuk setiap child class
  - Class Diagram-nya

```python
class anjing():
  #contructor
  def __init__(self,nama,jenis_kelamin,umur,ras,warna):
    self.nama = nama
    self.jenis_kelamin = jenis_kelamin
    self.umur = umur
    self.ras = ras
    self.warna = warna

  #method getter
  def get_info_anjing(self):
    return self.nama, self.jenis_kelamin, self.umur, self.ras, self.warna

  #method setter
  def set_info_anjing(self,nama,jenis_kelamin,umur,ras,warna):
    self.nama = nama
    self.jenis_kelamin = jenis_kelamin
    self.umur = umur
    self.ras = ras
    self.warna = warna

  def bersuara(self):
    print(f"{self.nama} bersuara guk-guk ")
  def bergerak(self):
    print(f"{self.nama} bergerak menggunakan kaki")
  def tidur(self):
    print(f"{self.nama} tidur di malam hari")

Bruno = anjing('Bruno', 'Jantan', 'Anjing Gembala', 3, 'Coklat')
print(Bruno.get_info_anjing())
Bruno.bersuara()
Bruno.tidur()
Bruno.bergerak()

name, gender, race, age, color = Bruno.get_info_anjing()

print(f"""{name} adalah seekor {race} berumur {age} tahun, berjenis kelamin {gender}, dan memiliki bulu berwarna {color}""")
```
