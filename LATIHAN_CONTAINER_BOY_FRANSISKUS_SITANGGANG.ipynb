```python
Matriks5 = np.array([[23,35,93,87],
                    [92,44,11,83],
                    [40,12,46,28],
                    [73,67,65,32]])
print(Matriks5,'\n')
#BAGIAN A
print("==== BAGIAN A ====")
print(Matriks5[1:2,1:4],'\n')

#BAGIAN B
print("\n==== BAGIAN B ====")
print(Matriks5[2:,1:3])

#BAGIAN C
print("\n==== BAGIAN C ====\n")
print(Matriks5[1:,1:])

#BAGIAN D
print("\n==== BAGIAN D ====")
print(Matriks5[1:3,:2])

#BAGIAN E
print("\n==== BAGIAN E ====")
print(Matriks5[:2,:3])

#BAGIAN F
print("\n==== BAGIAN F ====")
print(Matriks5[1:,1:2])
```
